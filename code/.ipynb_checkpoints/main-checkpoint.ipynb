підключимаємо різні модулі 

```python
import csv
import numpy as np
import random
from numpy.random import random_sample
from scipy.stats import rv_discrete  
```

зчитаємо дані з файлу "prob_XX.csv"

```python
prob = []
with open('../vars/prob_16.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:    
        prob.append(row[0])
        
print(prob)
```

розділимо зчитані дані на ймовірнісні розподіли ВТ і ключів

```python
PT_dist = prob[0].split(',')
for i in range(len(PT_dist)):
    PT_dist[i] = float(PT_dist[i])
    
print('Plaintext distribution: ', PT_dist)

keys_dist = prob[1].split(',')
for i in range(len(keys_dist)):
    keys_dist[i] = float(keys_dist[i])
    
print('\nKeys distribution: ', keys_dist)
```

зчитаємо дані з файлу "table_XX.csv", та перетворимо їх в "нормальний" вигляд

```python
table = []
with open('../vars/table_16.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:    
        table.append(row[0])
for i in range(len(table)):
    table[i] = table[i].split(',')
    for j in range(len(table[i])):
        table[i][j] = int(table[i][j])
        
print(np.matrix(table))
```

знайдемо розподіл шифротекстів таким чином:
$P(C) = \sum\limits_{\forall (M,k): E_k(M)=C}P(M,k)$, де $P(M,k) = P(M) \cdot P(k)$.

```python
CT_dist = []
for i in range(20):
    CT_dist.append(0)
    for key_index in range(20):
        for PT_index in range(20):
            if (table[key_index][PT_index] == i):
                CT_dist[i] += PT_dist[PT_index] * keys_dist[key_index]

#далі ми округлюємо, через особливості чисел з плаваючою точкою        
for i in range(20):
    CT_dist[i] = round(CT_dist[i], 3)

print('Ciphertext distribution: ', CT_dist)
```

знайдемо розподіл $P(M,C)=\sum\limits_{\forall k: E_k(M)=C}P(M,k)$

```python
comp_dist = [[0 for i in range(20)] for i in range(20)]
for CT_index in range(20):
    for PT_index in range(20):
        for key_index in range(20):
            if (table[key_index][PT_index] == CT_index):
                comp_dist[PT_index][CT_index] += PT_dist[PT_index] * keys_dist[key_index]

print('Ciphertext and plaintext compatible distribution:\n\n', np.matrix(comp_dist))
```

обрахуємо розподіл умовних імовірностей $P(C|M)$

```python
cond_dist = [[0 for i in range(20)] for i in range(20)]
for PT_index in range(20):
    for CT_index in range(20):
        cond_dist[PT_index][CT_index] = comp_dist[PT_index][CT_index] / CT_dist[CT_index]
        
for i in range(20):
    for j in range(20):
        cond_dist[i][j] = round(cond_dist[i][j], 3)
print('Conditional distribution P(M|C): \n\n', np.matrix(cond_dist))
```

запропонуємо детерміністичну вирішуючу функцію

```python
def det_func (CT):
    result_arr = []
    for PT_index in range(20):
        for key_index in range(20):
            if (table[PT_index][key_index] == CT):
                result_arr.append(PT_index)
    return result_arr[(CT ** 2) % len(result_arr)]

#for i in range(20):
#    print(det_func(i))
```

функція витрат для запропонованої детерміністичної вирішуючої функції:

```python
def losses_det_func(PT,CT):
    if (det_func(CT) == PT):
        return 1
    else:
        return 0
    
```

обрахуємо середні витрати для запропонованої детерміністичної вирішуючої функції

```python
average_losses = 0

for PT in range(20):
    for CT in range(20):
        average_losses += comp_dist[PT][CT] * losses_det_func(CT,PT)
        
print("Average losses for proposed deterministic function: ", round(average_losses, 5))
```

побудуємо оптимальну детерміністичну вирішуючу функцію

```python
def opt_det_func(CT):
    PT_prob_dict = {}
    for PT_index in range(20):
        PT_prob_dict[PT_index] = cond_dist[PT_index][CT]
    max_prob = 0
    result = []
    for i in range(20):
        if (PT_prob_dict[i] > max_prob):
            max_prob = PT_prob_dict[i]
            result = [i]
        elif (PT_prob_dict[i] == max_prob):
            result.append(i)
    return random.choice(result)
```

функція витрат для оптимальної детерміністичної вирішуючої функції

```python
def losses_opt_det_func(PT, CT):
    if (opt_det_func(CT) == PT):
        return 1
    else:
        return 0
```

обрахуємо середні витрати для оптимальної детерміністичної вирішуючої функції

```python
average_losses_opt = 0

for PT in range(20):
    for CT in range(20):
        average_losses_opt += comp_dist[PT][CT] * losses_opt_det_func(CT,PT)
print("Average losses for optimal deterministic function: ", average_losses_opt)
```

реалізуємо функцію втрат для стохастичної вирішуючої функції

```python
def losses_stoch_func(PT, CT):
    return 1 - matrix.tolist()[CT][PT]
```

знаходження оптимальної стохастичної вирішуючої функції

```python
def opt_stoch_func(CT):
    PT_prob_dict = {}
    for PT_index in range(20):
        PT_prob_dict[PT_index] = cond_dist[PT_index][CT]
    max_prob = 0
    result = []
    for i in range(20):
        if (PT_prob_dict[i] > max_prob):
            max_prob = PT_prob_dict[i]
            result = [i]
        elif (PT_prob_dict[i] == max_prob):
            result.append(i)
    return result, max_prob
```

побудуємо матрицю оптимальної стохастичної вирішуючої функції

```python
opt_stoch_matrix = [[0 for i in range(20)] for j in range(20)]

for i in range(20):
    a = opt_stoch_func(i)
    if (len(a[0]) == 1):
        opt_stoch_matrix[i][a[0][0]] = 1
    elif (len(a[0]) == 2):
        for j in range(2):
            opt_stoch_matrix[i][a[0][j]] = 1/2
    elif (len(a[0]) == 3):
        for j in range(3):
            opt_stoch_matrix[i][a[0][j]] = 1/3
    elif (len(a[0]) == 4):
        for j in range(4):
            opt_stoch_matrix[i][a[0][j]] = 1/4
            
print(np.matrix(opt_stoch_matrix))
```

функція втрат для оптимальної стохастичної вирішуючої функції

```python
def losses_opt_stoch_func(PT, CT):
    return 1 - opt_stoch_matrix[CT][PT]
```
