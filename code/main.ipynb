підключимаємо різні модулі 

```python
import csv
import numpy as np
```

зчитаємо дані з файлу "prob_XX.csv"

```python
prob = []
with open('../vars/prob_16.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:    
        prob.append(row[0])
        
print(prob)
```

розділимо зчитані дані на ймовірнісні розподіли ВТ і ключів

```python
PT_dist = prob[0].split(',')
for i in range(len(PT_dist)):
    PT_dist[i] = float(PT_dist[i])
    
print('Plaintext distribution: ', PT_dist)

keys_dist = prob[1].split(',')
for i in range(len(keys_dist)):
    keys_dist[i] = float(keys_dist[i])
    
print('\nKeys distribution: ', keys_dist)
```

зчитаємо дані з файлу "table_XX.csv", та перетворимо їх в "нормальний" вигляд

```python
table = []
with open('../vars/table_16.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in reader:    
        table.append(row[0])
for i in range(len(table)):
    table[i] = table[i].split(',')
    for j in range(len(table[i])):
        table[i][j] = int(table[i][j])
        
print(np.matrix(table))
```

знайдемо розподіл шифротекстів таким чином:
$P(C) = \sum\limits_{\forall (M,k): E_k(M)=C}P(M,k)$, де $P(M,k) = P(M) \cdot P(k)$.

```python
CT_dist = []
for i in range(20):
    CT_dist.append(0)
    for PT_index in range(20):
        for key_index in range(20):
            if (table[PT_index][key_index] == i):
                CT_dist[i] += PT_dist[PT_index] * keys_dist[key_index]

#далі ми округлюємо, через особливості чисел з плаваючою точкою        
for i in range(20):
    CT_dist[i] = round(CT_dist[i], 3)

print('Ciphertext distribution: ', CT_dist)
```

знайдемо розподіл $P(M,C)$

```python
table_1 = [[0 for i in range(20)] for i in range(20)]
for PT_index in range(20):
    for CT_index in range(20):
        for key_index in range(20):
            if (table[PT_index][key_index] == CT_index):
                table_1[PT_index][CT_index] += PT_dist[PT_index] * keys_dist[key_index]

print(np.matrix(table_1))
```
